```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv('abc_report.csv', sep=';')
df
```

```python
text = list((df['NamaMenu']+',')*df['Total'])
len(text)
```

```python
text[:2]
```

```python
list_text = []
textt = []

for i in text :
    i = i.split(',')
    list_text.append(i)
        
for i in list_text:
    for j in i:
        j = j.replace(' ','')
        j = j.replace('Add','')
        if len(j)==0:
            continue
        else:
            textt.append(j)
```

```python
len(textt)
```

```python
myset = set(textt)  #cek aja distinct value nya apa aja
```

```python
str =  ' '.join(textt)
```

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, 
                      background_color='salmon', colormap='Pastel1', collocations=False).generate(str)
# Plot

plt.figure(figsize=(10,15))
plt.imshow(wordcloud)
plt.margins(x=0, y=0)
plt.axis("off")
```

```python
namaMenu = df.groupby(by=["NamaMenu"]).sum()
namaMenu
```

```python
sns.scatterplot(data=namaMenu, x='Total', y='TotalHarga')
```

```python
sns.regplot(data=namaMenu, x='Total', y='TotalHarga')
```

```python
harga_menuu = df[['NamaMenu','Harga']]
```

```python
harga_menuu.drop_duplicates()
```

```python
harga_menu = pd.merge(namaMenu,  
                      harga_menuu,  
                      on ='NamaMenu',  
                      how ='inner') 
harga_menu = harga_menu.drop_duplicates()
```

```python
sns.scatterplot(data=harga_menu, x='Total', y='Harga_y')
```

```python
sorted10 = harga_menu.sort_values('TotalHarga', ascending=False)
```

```python
top10 = harga_menu.sort_values('TotalHarga', ascending=False).head(10)
top10
```

```python
sns.barplot(data=top10, x='TotalHarga', y='NamaMenu')
```

```python
result = sorted10.nlargest(5, columns='TotalHarga')
result
```

```python
sorted10.loc[sorted10.NamaMenu.isin(result.NamaMenu), 'TotalHarga'].sum()
```

```python
sorted10.loc[~sorted10.NamaMenu.isin(result.NamaMenu), 'TotalHarga'].sum()
```

```python
dictt = {'sum' : [sorted10.loc[sorted10.NamaMenu.isin(result.NamaMenu), 'TotalHarga'].sum(), sorted10.loc[~sorted10.NamaMenu.isin(result.NamaMenu), 'TotalHarga'].sum()]}
dictt
```

```python
dictt1 = pd.DataFrame(dictt)
dictt1['type'] = ['top5','rest_of_menu']
dictt1
```

```python
plt.figure(figsize=(10,5))

values = dictt1['sum']
labels = dictt1['type']
plt.pie(values, labels = labels, autopct='%1.1f%%')
plt.title('Percentage by Severity Description')
plt.tight_layout()
plt.show
```
